## 연도별 전체 진료수

```python
import pandas as pd
df_month=pd.read_csv("비염_연도별진료수.csv")
```

```python
df_month.head()
```

```python
df_month = df_month.rename(columns={'합계 : 진료에피소드 건수':'진료수'})
```

```python
df_month["년도"]=df_month["요양개시년월"].str[0:6]
df_month["요약"]=df_month["요양개시년월"].str[7:]
```

```python
df_month
```

```python
df_month
```

```python
df_month=df_month.drop(['요양개시년월'],axis=1)
df_month=df_month.drop(['요약'],axis=1)
```

```python
df_month
```

```python
df_month["월"]=df_month["년도"].str[4:]
df_month["년도"]=df_month["년도"].str[0:4]
```

```python
df_month
```

```python
df_year=df_month.groupby(['년도'], as_index=False).sum()
```

```python
df_year
```

```python
df_year=df_year.drop(['월'],axis=1)
```

```python
df_year
```

```python
df_year.to_csv('/Users/yeji/SWING 2023 해커톤 with LONDON/전처리데이터/연도별/비염연도별.csv') #csv파일로 생성
```

```python
import numpy as np
import matplotlib.pyplot as plt
#연도별 진료수 그래프

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

df_year.plot.bar(x='년도', y='진료수', rot=0,legend=False)
plt.title('12개년 비염 진료수')
plt.xlabel('년도')
plt.ylabel('진료수 합')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#연도별 진료수 그래프 ( y축 2000000부터 시작)

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

df_year.plot.bar(x='년도', y='진료수', rot=0,legend=False)
plt.title('12개년 비염 진료수')
plt.ylim([2000000, 5000000])
plt.xlabel('년도')
plt.ylabel('진료수 합')
plt.show()
```

## 비염 연령군5 연도별 진료수

```python
import pandas as pd
five_month_df=pd.read_csv("비염_5연령군_연도별진료수.csv")
```

```python
five_month_df
```

```python
five_month_df= five_month_df.rename(columns={'합계 : 진료에피소드 건수':'진료수'})
```

```python
five_month_df
```

```python
five_month_df= five_month_df.rename(columns={'행 레이블':'년도월'})
```

```python
five_month_df
```

```python
five_month_df = five_month_df.astype({'년도월':'string'})
five_month_df
```

```python
five_month_df["월"]=five_month_df["년도월"].str[4:]
five_month_df["년도"]=five_month_df["년도월"].str[0:4]
```

```python
five_month_df
```

```python
five_month_df=five_month_df.drop(['월'],axis=1)
```

```python
five_month_df=five_month_df.drop(['년도월'],axis=1)
```

```python
five_month_df
```

```python
five_month_df=five_month_df.groupby(['년도'], as_index=False).sum()
```

```python
five_month_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
#연도별 진료수 그래프

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

five_month_df.plot.bar(x='년도', y='진료수', rot=0,legend=False)
plt.title('12개년 연령군 5의 비염 진료수')
plt.xlabel('년도')
plt.ylabel('진료수 합')
plt.show()
```
