```python
import pandas as pd
df=pd.read_csv("주의보_경보발령내역(2010-2021).csv")
df2_month=pd.read_csv("[2010-2021]전체_대기오염지수.csv")
import numpy as np
import matplotlib.pyplot as plt

df.head()
```

```python
df
```

```python
df_year=df.groupby(['년도'], as_index=False).value_counts()
```

```python
df2_year=df2_month.groupby(['측정년도'], as_index=False).mean()
```

```python
df_year
```

```python
df2_year
```

```python
df_count=df_year["count"].groupby([df_year['물질명'],df_year['년도']]).sum()
```

```python
df_count=df_count.to_frame()
```

```python
df_count
```

```python
#특정 행의 열 값 가져오는 테스트
df_sample=df_count.loc[['미세먼지']]
df_sample
```

```python
#미세먼지, 초미세먼지 막대그래프 그리기
#데이터프레임 생성
df = pd.DataFrame([[1,0,0,]],
                  index=['달리기','점프'],
                 columns=['모모','츄츄','디디','루루'])
df
```

```python
#미세먼지, 초미세먼지 막대그래프 그리기
x = np.arange(10)
years = ['2010', '2013', '2014','2015','2016','2017','2018','2019','2020','2021']
values = [1,1,2,3,6,6,6,8,4,10]

plt.bar(x, values)
plt.xticks(x, years)

plt.title('PM10.0 Warning')
plt.xlabel('YEARS')
plt.ylabel('COUNTS')

plt.show()

```

```python
#미세먼지 주의보, 평균 그래프 겹쳐 시각화

#미세먼지 평균 막대그래프
x = np.arange(2010,2022)

Bs = df2_year['미세먼지']

plt.bar(x, Bs,label='Density Average')

ax = plt.subplot()
ax.set_xticks(x)
#ax.set_xticklabels(df2_year['미세먼지'])

#미세먼지 주의보 선그래프
#years = ['2010', '2013', '2014','2015','2016','2017','2018','2019','2020','2021']
values = [1,0,0,1,2,3,6,6,6,8,4,10]
plt.plot(x, values, label='Warning Count')


plt.title('PM10.0 Density Average-Warning Count')
plt.xlabel('YEARS')
plt.ylabel('COUNTS')

plt.legend()
plt.show()
```
