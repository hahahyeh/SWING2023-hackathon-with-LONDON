```python
import matplotlib.font_manager as fm
topics=df_year['측정년도']
Cs = df_year['일산화탄소']
Ds = df_year['오존']
Fs = df_year['이산화질소']
Es = df_year['아황산가스']

f_bottom = np.add(Cs, Ds)
e_bottom = np.add(f_bottom, Es)

x = range(len(topics))


plt.bar(x, Cs)
plt.bar(x, Ds, bottom=Cs)

plt.bar(x, Fs, bottom=f_bottom)
plt.bar(x, Es, bottom=e_bottom)

ax = plt.subplot()
ax.set_xticks(x)
ax.set_xticklabels(topics)
plt.title('10개년 대기오염물질 평균 추이')
plt.xlabel('YEAR')
plt.ylabel('DENSITY')
plt.rcParams['font.family'] = 'LG PC'
plt.show()
```
