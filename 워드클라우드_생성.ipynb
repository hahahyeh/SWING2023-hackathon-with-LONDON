```python
import os
import pandas as pd
import numpy as np
# os.getcwd() #현재 경로 확인
# os.chdir(r"/Users/yeji/SWING 2023 해커톤 with LONDON")
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
df = pd.read_csv("/content/환경카테고리_기사제목크롤링_최종.csv", engine="python")
```

```python
df
```

```python
text = ''
for i in range(20303):
    text = text + df.iloc[i, 2]
```

```python
okt = Okt()
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
```

```python
wc = WordCloud(font_path='/content/drive/MyDrive/NanumGothic.otf', width=800, height=800, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
plt.savefig('환경카테고리_워드클라우딩4.png')
```
