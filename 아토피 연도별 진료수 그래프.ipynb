## 아토피 전체연령 연도별 진료수

```python
import pandas as pd
month_df=pd.read_csv("아토피_전체연령군_연월별진료수.csv")
```

```python
month_df
```

```python
month_df = month_df.astype({'년월':'string'})
month_df
```

```python
month_df["월"]=month_df["년월"].str[4:]
month_df["년도"]=month_df["년월"].str[0:4]
```

```python
month_df
```

```python
month_df=month_df.drop(['월'],axis=1)
```

```python
month_df=month_df.drop(['년월'],axis=1)
```

```python
month_df
```

```python
month_df=month_df.groupby(['년도'], as_index=False).sum()
```

```python
month_df
```

```python
month_df.to_csv('/Users/yeji/SWING 2023 해커톤 with LONDON/전처리데이터/연도별/전체아토피연도별.csv') #csv파일로 생성
```

```python
import numpy as np
import matplotlib.pyplot as plt
#연도별 진료수 그래프

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

month_df.plot.bar(x='년도', y='진료수', rot=0,legend=False)
plt.title('12개년 아토피 진료수')
plt.xlabel('년도')
plt.ylabel('진료수 합')
plt.show()
```
