```python
import pandas as pd
df=pd.read_csv("경보발령수(년도,발령수).csv")
import numpy as np
import matplotlib.pyplot as plt

df.head()
```

```python
plt.plot(df['년도'],df['발령수'],color='blue',linestyle='-',marker='o')

plt.xlabel('YEAR') #x 라벨
plt.ylabel('COUNT') #y 라벨
plt.title("Air Pollution Warning Count") #그래프 이름

plt.legend(['Warning Count'])

#그래프 출력
plt.show()
```
